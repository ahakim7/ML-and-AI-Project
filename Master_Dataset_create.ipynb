Data Upload

```python
import pandas as pd

# Load all datasets
flic_df = pd.concat([
    pd.read_csv("flic_button_data_export_20251110.csv"),
    pd.read_csv("flic_button_data_export_20251127.csv"),
    pd.read_csv("flic_button_data_export_20260109.csv")
])

g_df = pd.concat([
    pd.read_csv("g_data_export_20251110.csv"),
    pd.read_csv("g_data_export_20251127.csv"),
    pd.read_csv("g_data_export_20260109.csv")
])

md_df = pd.concat([
    pd.read_csv("md30_export_20251110.csv"),
    pd.read_csv("md30_export_20251127.csv"),
    pd.read_csv("md30_export_20260109.csv")
])
```

Time Date formate fix

```python
for df in [flic_df, g_df, md_df]:
    df["timestamp"] = pd.to_datetime(df["ts"], utc=True, errors="coerce")
    df["timestamp"] = df["timestamp"].dt.round("1s")

flic_df["source"] = "flic"
g_df["source"] = "gdata"
md_df["source"] = "md30"
```

Combine Dataset

```python
# First merge Flic + G-data
merged = pd.merge(flic_df, g_df, on="timestamp", how="outer", suffixes=("_flic", "_g"))

# Then merge MD30
merged = pd.merge(merged, md_df, on="timestamp", how="outer", suffixes=("", "_md30"))
```

Sort and Save Master Dataset

```python
merged = merged.sort_values("timestamp")

merged.to_csv("MASTER_HAZARD_DATASET.csv", index=False)
```
